```python
import numpy as np
```

## `np.diag()`

```python
np.diag([1,2,3])
```

```python
np.diag([1,2,3], k=1) # k - the diagonal. 0 = main diagonal, k>0 for diagonals ABOVE the main
```

```python
np.diag([1,2,3], k=3)
```

```python
np.diag([1,2,3], k=-1) # and uses k<0 for diagonals BELOW the main
```

## `np.indices()`

```python
ind = np.indices((8,8))
```

```python
print(ind.shape)
```

```python
print(ind)
```

```python
parity = ind & 1
```

```python
print(parity)
```

```python
parity.shape
```

```python
parity.sum(axis=0) # collapse axis 0 by summing along it
```
