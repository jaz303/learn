#### 22. Normalize a 5x5 random matrix (★☆☆)

````python
vals = np.random.rand(5,5)

# sum to 1
norm = vals / np.sum(vals)
print(norm.sum())

# scale to max 1
norm = vals / np.max(vals)
print(norm)

# scale to range [0,1]
norm = (vals - vals.min()) / (vals.max() - vals.min())
print(norm)
````

#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
# attempt 1 - clobbers A

A = np.array([[1.,2.], [3.,4.]])
B = np.array([[5.,6.], [6.,7.]])

Out = np.copy(A)
Out += B
A *= -0.5
Out *= A

print(Out)

# attempt 2 - rearrange formula and we can avoid clobbering A

A = np.array([[1.,2.], [3.,4.]])
B = np.array([[5.,6.], [6.,7.]])

Out = np.copy(A)
Out += B
Out *= -0.5
Out *= A
print(Out)

# attempt 3 - if we're ok with clobbering, we can do it without an Out variable
# and use B as the destination. this only works because B is only referenced once!

A = np.array([[1.,2.], [3.,4.]])
B = np.array([[5.,6.], [6.,7.]])

B += A
B *= -0.5
B *= A
print(B)
````
